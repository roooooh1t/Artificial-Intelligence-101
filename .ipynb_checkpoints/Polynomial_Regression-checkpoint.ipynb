## Demonstration of Underfitting , Proper Fitting and Overfitting using Polynomial Regression

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

### Data Generation

```python
np.random.seed(42)
x = np.linspace(-10 , 10 , 100)
noise_std = 10
y = (0.5 * x**3) - (2 * x**2) + (3 * x) + np.random.normal(0 , noise_std , size = x.shape)
```

### Polynomial Fits

```python
#Degree of the curve
underfit_degree = 1
properfit_degree = 3
overfit_degree = 29

# Fit polynomial models of specified degrees
underfit_coeffs = np.polyfit(x, y, underfit_degree)
properfit_coeffs = np.polyfit(x, y, properfit_degree)
overfit_coeffs = np.polyfit(x, y, overfit_degree)

# Evaluate polynomial models on x
y_underfit = np.polyval(underfit_coeffs, x)
y_properfit = np.polyval(properfit_coeffs, x)
y_overfit = np.polyval(overfit_coeffs, x)
```

### Plotting Results

```python
fig, axes = plt.subplots(1 , 3 , figsize = (18 , 6))

# Underfit model
axes[0].scatter(x , y , color = 'black' , label = 'Data' , alpha = 0.6)
axes[0].plot(x , y_underfit , label = f'Underfit (Degree {underfit_degree})' , color = 'red' , linestyle = 'dashed')
axes[0].set_title('Underfitting')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()
axes[0].grid(True)

# Proper Fit model
axes[1].scatter(x , y , color = 'black' , label = 'Data' , alpha = 0.6)
axes[1].plot(x , y_properfit , label = f'Proper Fit (Degree {properfit_degree})' , color = 'green', linestyle = 'solid')
axes[1].set_title('Proper Fitting')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].legend()
axes[1].grid(True)

# Overfit model
axes[2].scatter(x , y , color = 'black', label = 'Data', alpha=0.6)
axes[2].plot(x, y_overfit, label = f'Overfit (Degree {overfit_degree})', color = 'blue', linestyle = 'dotted')
axes[2].set_title('Overfitting')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()
```
