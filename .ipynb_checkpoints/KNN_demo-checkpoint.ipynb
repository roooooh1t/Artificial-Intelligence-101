## k-NN Classifier

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Custom kNN

```python
def my_knn_classifier(train_data , train_labels , test_data , k):
    """
    Custom k-Nearest Neighbors classifier.

    Parameters:
        train_data (ndarray): Feature matrix for training data.
        train_labels (ndarray): Labels corresponding to training data.
        test_data (ndarray): Feature matrix for test points.
        k (int): Number of neighbors to consider.

    Returns:
        ndarray: Predicted labels for each test point.
    """
    predictions = []

    for test_point in test_data:
        # Step 1: Compute distances from test_point to all training points
        distances = np.linalg.norm(train_data - test_point , axis = 1)

        # Step 2: Find the indices of the k smallest distances
        k_nearest_indices = np.argsort(distances)[:k]

        # Step 3: Get the labels of the k nearest neighbors
        k_nearest_labels = train_labels[k_nearest_indices]

        # Step 4: Count the frequency of each label
        labels, counts = np.unique(k_nearest_labels, return_counts=True)

        # Step 5: Pick the label with the highest count (majority vote)
        most_common_label = labels[np.argmax(counts)]

        # Step 6: Save the predicted label
        predictions.append(most_common_label)

    return np.array(predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_knn_classifier(train_data, train_labels, test_data, k):
    """
    Custom k-Nearest Neighbors classifier.

    Parameters:
        train_data (ndarray): Feature matrix for training data.
        train_labels (ndarray): Labels corresponding to training data.
        test_data (ndarray): Feature matrix for test points.
        k (int): Number of neighbors to consider.

    Returns:
        ndarray: Predicted labels for each test point.
    """
    predictions = []

    # Loop through each point we want to classify
    for point in test_data:
        # Calculate Euclidean distances from this test point to all training points
        dists = np.linalg.norm(train_data - point, axis=1)

        # Get the indices of the k closest points
        top_k_indices = np.argsort(dists)[:k]

        # Get the labels of the k nearest neighbors
        top_k_labels = train_labels[top_k_indices]

        # Count occurrences of each label among the neighbors
        labels, label_counts = np.unique(top_k_labels, return_counts=True)

        # Choose the label that appears most frequently
        predicted_label = labels[np.argmax(label_counts)]
        predictions.append(predicted_label)

    return np.array(predictions)


np.random.seed(42)  # For reproducibility

num_points = 100  # Number of samples per class

# Define means for the two classes
mean_a = [1.5, 1.5]
mean_b = [-2.0, -2.0]

# Define covariance (controls shape & orientation of clusters)
cov_a = [[0.9, 0.3], [0.3, 0.9]]
cov_b = [[1.0, -0.5], [-0.5, 1.0]]

# Sample points from 2D Gaussian distributions
points_a = np.random.multivariate_normal(mean_a, cov_a, num_points)
points_b = np.random.multivariate_normal(mean_b, cov_b, num_points)

# Combine into one dataset
features = np.vstack((points_a, points_b))
labels = np.hstack((np.zeros(num_points), np.ones(num_points)))  # 0 for class A, 1 for class B

plt.scatter(points_a[:, 0], points_a[:, 1], color='green', marker='o', label='Class A')
plt.scatter(points_b[:, 0], points_b[:, 1], color='red', marker='^', label='Class B')

# Define range of x and y for mesh grid
x_start, x_end = features[:, 0].min() - 1, features[:, 0].max() + 1
y_start, y_end = features[:, 1].min() - 1, features[:, 1].max() + 1

# Create a grid of points across the feature space
grid_x, grid_y = np.meshgrid(np.arange(x_start, x_end, 0.02),
                             np.arange(y_start, y_end, 0.02))

# Stack grid into (N, 2) shape for classification
grid_points = np.c_[grid_x.ravel(), grid_y.ravel()]

# Choose k and predict the class for each grid point
k_value = 3  # Try different k values like 1, 3, 5, etc.
grid_predictions = my_knn_classifier(features, labels, grid_points, k_value)
grid_predictions = grid_predictions.reshape(grid_x.shape)

# Plot decision regions
plt.contourf(grid_x, grid_y, grid_predictions, alpha=0.3, cmap='coolwarm')

# Final plot details
plt.title(f'k-NN Classification (k = {k_value})')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def knn_regression(X_train, y_train, X_test, k):
    """
    k-Nearest Neighbors (kNN) regression.

    Parameters:
        X_train (ndarray): Training data features.
        y_train (ndarray): Training data labels (target values).
        X_test (ndarray): Test data features.
        k (int): Number of neighbors.

    Returns:
        ndarray: Predicted target values for the test data.
    """
    predicted_values = []

    for sample in X_test:
        distances = np.abs(X_train - sample)
        nearest_indices = np.argsort(distances)[:k]
        nearest_values = y_train[nearest_indices]
        predicted_values.append(np.mean(nearest_values))

    return np.array(predicted_values)

# Example usage and demo
# Generate synthetic 1D data points for regression
np.random.seed(0)
num_samples = 50
X_train = np.sort(5 * np.random.rand(num_samples))
y_train = np.sin(X_train) + 0.1 * np.random.randn(num_samples)  # True function with noise
X_test = np.linspace(0, 5, 100)

# Apply k-Nearest Neighbors (kNN) regression
k = 2  # Play with this value to see how the learned regression function's shape changes (note: for regresssion, it is okay to choose even value of k)
predicted_values = knn_regression(X_train, y_train, X_test, k)

# Plot the original data and the kNN regression curve
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_test, np.sin(X_test), color='green', label='True Function')
plt.plot(X_test, predicted_values, color='red', label=f'kNN Regression (k={k})')

plt.xlabel('Input')
plt.ylabel('Output')
plt.title('k-Nearest Neighbors (kNN) Regression')
plt.legend()
plt.grid(True)
plt.show()
```
